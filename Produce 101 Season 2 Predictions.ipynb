```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
full_df = pd.read_csv("produce101.csv")
season2_df = pd.read_csv("season2.csv")
```

```python
name1=[]

for name in season2_df['Name']:
    name = name.rsplit('(', 1)[0]
    name1.append(name)

season21_df = pd.DataFrame()
season21_df = season21_df.append(name1)
season2_df['Name_1'] = season21_df
```

```python
full_df = full_df[pd.notnull(full_df['Final_Rank'])]
```

```python
#Pre-processing of data 

full_df['Ep8_Rank'] = full_df['Ep8_Rank'].fillna(0)
full_df['Ep8_Votes'] = full_df['Ep8_Votes'].fillna(0)
full_df['Ep10_Rank'] = full_df['Ep10_Rank'].fillna(0)
full_df['Ep10_Votes'] = full_df['Ep10_Votes'].fillna(0)
full_df['Ep11_Rank'] = full_df['Ep11_Rank'].fillna(0)
full_df['Ep11_Votes'] = full_df['Ep11_Votes'].fillna(0)
full_df['Evaluation1'] = full_df['Evaluation1'].fillna(0)
full_df['Evaluation2'] = full_df['Evaluation2'].fillna(0)
full_df['Ep1_Rank'] = full_df['Ep1_Rank'].fillna(full_df['Ep1_Rank'].mean())
full_df['Ep2_Rank'] = full_df['Ep2_Rank'].fillna(full_df['Ep2_Rank'].mean())
full_df['Ep3_Rank'] = full_df['Ep3_Rank'].fillna(full_df['Ep3_Rank'].mean())
full_df['Ep5_Rank'] = full_df['Ep5_Rank'].fillna(full_df['Ep5_Rank'].mean())
full_df['Ep5_Votes'] = full_df['Ep5_Votes'].fillna(full_df['Ep5_Votes'].mean())

full_df['is_company'] = np.where(full_df['Company'] == "Individual Trainee_(__ ___)", 1, 0)
```

```python
#Pre-processing of data 


season2_df['Evaluation1'] = season2_df['Evaluation1'].fillna(0)
season2_df['Evaluation2'] = season2_df['Evaluation2'].fillna(0)
season2_df['Ep1_Rank'] = season2_df['Ep1_Rank'].fillna(season2_df['Ep1_Rank'].mean())
season2_df['Ep2_Rank'] = season2_df['Ep2_Rank'].fillna(season2_df['Ep2_Rank'].mean())
season2_df['Ep3_Rank'] = season2_df['Ep3_Rank'].fillna(season2_df['Ep3_Rank'].mean())
season2_df['Ep4_Rank'] = season2_df['Ep4_Rank'].fillna(season2_df['Ep4_Rank'].mean())
season2_df['Ep4_Votes'] = season2_df['Ep4_Votes'].fillna(season2_df['Ep4_Votes'].mean())
```

```python

from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
full_df['Evaluation1_New'] = labelencoder_X.fit_transform(full_df['Evaluation1'])
full_df['Evaluation2_New'] = labelencoder_X.fit_transform(full_df['Evaluation2'])

#preprocessing for season 2 data
season2_df['Evaluation1_New'] = labelencoder_X.fit_transform(season2_df['Evaluation1'])
season2_df['Evaluation2_New'] = labelencoder_X.fit_transform(season2_df['Evaluation2'])
```

```python
#Regression starts here
cols_to_use = ['Evaluation1_New', 'Evaluation2_New', 'Ep1_Rank', 'Ep2_Rank', 'Ep3_Rank', 'Ep5_Rank', 'Ep5_Votes']

X = full_df[cols_to_use]
y = full_df['Final_Rank']

#Season 2 to use

season2_cols_to_use = ['Evaluation1_New', 'Evaluation2_New', 'Ep1_Rank', 'Ep2_Rank', 'Ep3_Rank', 'Ep4_Rank', 'Ep4_Votes']

season_test = season2_df[season2_cols_to_use]


# Splitting the dataset into the Training set and Test set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 300, random_state = 0)
regressor.fit(X_train, y_train)

```

```python
y_pred = regressor.predict(season_test)


acc_decision_tree = round(regressor.score(X_test, y_test) * 100, 2)
print "accuracy: %r" %(acc_decision_tree) + "%"
```

```python
#Creating new index 

submission = pd.DataFrame({
        "Name": season2_df['Name_1'],
        "Ranking": y_pred
    })
```

```python
submission = submission.sort(['Ranking'], ascending=[True])
submission = submission.reset_index(drop=True)

submission 
```
